```python
import paramiko
import time
import sys

ip = "172.16.16.122"
user = "root"
pw = "asd123!@"

# ---- 원격 실행 및 실시간 출력 함수 ----
def run_remote(script):
    ssh = paramiko.SSHClient()
    ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    ssh.connect(ip, username=user, password=pw, port=22)

    shell = ssh.invoke_shell()
    shell.send(script + "\n")

    while True:
        if shell.recv_ready():
            output = shell.recv(4096).decode('utf-8', errors='ignore')
            print(output, end='')
            sys.stdout.flush()
        # exit_status_ready는 단순 명령에는 바로 True가 되지 않으므로, timeout 기반 루프
        if shell.closed:
            break
        time.sleep(0.1)

    ssh.close()


bash_script = """
set -e

echo "1. 네트워크 및 기본 설정 확인"
ip -4 addr show | grep -oP '(?<=inet\\s)\\d+(\\.\\d+){3}'
ping -c 3 172.16.0.1 || echo "게이트웨이 응답 없음 (무시)"

echo "2. DNS 확인"
cat /etc/resolv.conf

echo "3. Repository 설정"
dnf install -y epel-release
dnf install -y https://rpms.remirepo.net/enterprise/remi-release-9.rpm
dnf -y update && dnf -y upgrade

echo "4. 필수 도구 설치"
dnf install -y vim net-tools wget
dnf install -y expect

echo "5. 방화벽 및 SELinux 해제"
sed -i 's/^SELINUX=.*/SELINUX=disabled/' /etc/selinux/config
systemctl disable --now firewalld || true
echo "※ SELinux 비활성화를 위해 재부팅 필요할 수 있음"

echo "6. Miniconda 자동 설치"

# Miniconda 설치 파일 다운로드
wget https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh -O /root/Miniconda3-latest-Linux-x86_64.sh

# Expect 스크립트 작성
cat <<'EOF' > /root/miniconda_expect.sh
#!/usr/bin/expect -f

spawn bash /root/Miniconda3-latest-Linux-x86_64.sh
expect "Do you accept the license terms?"
send "yes\r"
expect "Press ENTER to continue"
send "\r"
expect -re "(initialize|update).*conda.*?\[yes\|no\]"
send "no\r"
expect eof
EOF




# 실행 권한 부여
chmod +x /root/miniconda_expect.sh

# 자동 설치 실행
/usr/bin/expect /root/miniconda_expect.sh


echo "7. PATH 등록"
echo 'export PATH=/root/miniconda3/bin:$PATH' >> /etc/bashrc
source /etc/bashrc
source /etc/bashrc

echo "8. Jupyter Notebook 설치"

conda tos accept --override-channels --channel https://repo.anaconda.com/pkgs/main
conda tos accept --override-channels --channel https://repo.anaconda.com/pkgs/r
conda install -y jupyter

echo "9. 주피터 설정 생성 및 비밀번호 설정"


cat <<'EOF' > /root/.jupyter/jupyter_notebook_config.py
c.ServerApp.ip = '172.16.16.122'
c.ServerApp.port = 8000
c.ServerApp.root_dir = '/root/'
c.ServerApp.open_browser = False
c.ServerApp.password = 'asd123!@'
EOF

jupyter notebook --generate-config

echo "10. Jupyter 서비스 등록"
cat <<'EOF' > /etc/systemd/system/jupyter.service
[Unit]
Description=Jupyter Notebook Server
After=network.target

[Service]
Type=simple
User=root
Environment="PATH=/root/miniconda3/bin:/usr/local/sbin:/usr/local/bin:/usr/sbin:/usr/bin"
ExecStart=/root/miniconda3/bin/jupyter-notebook --allow-root --config=/root/.jupyter/jupyter_notebook_config.py
Restart=always

[Install]
WantedBy=multi-user.target
EOF


echo "11. Jupyter 서비스 활성화 및 자동 시작"
systemctl daemon-reload
systemctl enable --now jupyter

echo "12. 서비스 상태 확인"
systemctl status jupyter.service

echo "✅ 모든 설정 완료!"
echo "브라우저에서 접속: http://172.16.16.122:8000"
"""

# 실행
run_remote(bash_script)
```
