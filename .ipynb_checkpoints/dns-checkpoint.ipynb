```python
import paramiko
import time
import sys
import os

ip = "172.16.16.122"
user = "root"
pw = "asd123!@"

def cc(cmd):
    """SSH로 명령을 실행하고 결과를 출력"""
    cli = paramiko.SSHClient()
    cli.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    cli.connect(ip, username=user, password=pw, port=22)
    stdin, stdout, stderr = cli.exec_command(cmd)
    out = stdout.read().decode()
    err = stderr.read().decode()
    cli.close()
    if err:
        print("⚠️ 오류 발생:", err)
    return out.strip()

def install() :
    cc( 'echo "=== install ===" ')
    cc("dnf -y install bind bind-utils")
    cc("systemctl status named")

def modify_named_conf_file() :
    #!/bin/bash
    cmd = r"""
    NAMED_CONF="/etc/named.conf"
    sed -i 's/listen-on[[:space:]]*port 53[[:space:]]*{[[:space:]]*127\.0\.0\.1;[[:space:]]*};/listen-on port 53 { any; };/' "$NAMED_CONF"
    sed -i 's/allow-query[[:space:]]*{[[:space:]]*localhost;[[:space:]]*};/allow-query { any; };/' "$NAMED_CONF"
    """
    print(cc(cmd))
    print("✅ named.conf 수정 완료")

def modify_named_rfc_file() :

    cmd = r"""
    NAMED_CONF="/etc/named.rfc1912.zones"

    cat >> "$NAMED_CONF" <<'EOF'

zone "team1.com" IN {
        type master;
        file "team1.com";
        allow-update { none; };
};
EOF
    """
    print(cc(cmd))
    print("✅ zone 'team1.com' 설정이 추가되었습니다.")

def add_namezone() :
    cc("cp -ap /var/named/named.localhost /var/named/team1.com.zone")

#####################################################
##             dns_data.txt 생성코드                ##
#####################################################


## 여기 부분부터는 원격으로는 못함.. ##
## 1. dns 생성 파일 
def create_dns_file() :
    file_name = "dns_data.txt"
    target_path = "../output/"
    command1 = "touch %s%s"%(target_path, file_name)
    a = os.popen(command1)
    data = a.read()

# ================================  #

## 2. dns 파일 추가 생성 코드
def info():
    name = input("이름을 입력하세요: ")
    domain_id = input("사용할 도메인 이름을 입력하세요: ")
    ip = input("IP를 입력하세요: ")

    # 원하는 포맷으로 문자열 생성
    data = f"{name} {domain_id} {ip}"
    return data

def dns_mod() :
    while True:
    # 파일 읽기
    with open(file_path, "r") as f:
        content = f.read()

    lines = content.splitlines()
    print("\n==== 현재 파일 내용 ====")
    for i, line in enumerate(lines, 1):
        print(f"{i}: {line}")
    print("======================\n")

    # 사용자 입력
    input_num = input("수정(1), 삭제(2), 추가(3), 종료(0) 원하시는 동작의 번호를 입력하세요: ")

    if input_num == "0":
        print("프로그램 종료")
        break

    elif input_num == "1":  # 수정
        line_num = int(input("수정할 줄 번호를 입력하세요: "))
        if 1 <= line_num <= len(lines):
            # info() 함수 호출해서 새로운 데이터 받기
            new_text = info()
            lines[line_num - 1] = new_text
            with open(file_path, "w") as f:
                f.write("\n".join(lines))
        else:
            print("잘못된 줄 번호입니다.")
    
    elif input_num == "2":  # 삭제
        line_num = int(input("삭제할 줄 번호를 입력하세요: "))
        if 1 <= line_num <= len(lines):
            lines.pop(line_num - 1)
            # 빈 줄 제거
            lines = [line for line in lines if line.strip() != ""]
            with open(file_path, "w") as f:
                f.write("\n".join(lines))
        else:
            print("잘못된 줄 번호입니다.")

    elif input_num == "3":  # 추가
        new_text = info()
        with open(file_path, "a") as f:
            f.write("\n" + new_text)

    else:
        print("잘못된 선택입니다. 0~3 사이의 숫자를 입력하세요.")

## dns 파일을 덮어쓰기
def overwrite_dns() :
    zone_file = "/var/named/team1.com.zone"
    data_file = "../output/dns_data.txt"
    
    data_lines = []
    zone_lines = []
    
    
    # 취합된 파일 읽기
    with open(data_file, "r", encoding="utf-8") as d:
        data_lines = d.readlines()
    
    # Zone 파일 읽기
    with open(zone_file, "r", encoding="utf-8") as db:
        zone_lines = db.readlines()
    
    # Zone 파일의 12번째 줄까지 유지하고 이후는 삭제
    header = zone_lines[:16]
    
    # 취합된 데이터를 규칙에 맞게 변환
    new_records = []
    for line in data_lines:
        parts = line.strip().split()
        if len(parts) >= 3:
            name, c3, ip = parts[:3]
            new_records.append(f"{c3}\tIN\tA\t{ip}\n\n")
    
    # 합치기
    new_zone_content = header + new_records
    
    # Zone 파일 덮어쓰기
    with open(zone_file, "w", encoding="utf-8") as f:
        f.writelines(new_zone_content)
    
    print("Zone 파일이 성공적으로 갱신되었습니다.")

    

def init() :
#    install()
#    modify_named_conf_file()
#    modify_named_rfc_file()
    add_namezone()
    create_dns_file()

def mod_zone() :
    dns_mod()
    overwrite_dns()

init()
mod_zone()
```
