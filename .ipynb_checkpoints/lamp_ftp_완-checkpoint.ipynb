```python
import paramiko
import time
import sys

ip = "172.16.16.122"
user = "root"
pw = "asd123!@"

# ---- ì›ê²© ì‹¤í–‰ ë° ì‹¤ì‹œê°„ ì¶œë ¥ í•¨ìˆ˜ ----
def run_remote(script):
    ssh = paramiko.SSHClient()
    ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    ssh.connect(ip, username=user, password=pw, port=22)

    shell = ssh.invoke_shell()
    shell.send(script + "\n")

    while True:
        if shell.recv_ready():
            output = shell.recv(4096).decode('utf-8', errors='ignore')
            print(output, end='')
            sys.stdout.flush()
        # exit_status_readyëŠ” ë‹¨ìˆœ ëª…ë ¹ì—ëŠ” ë°”ë¡œ Trueê°€ ë˜ì§€ ì•Šìœ¼ë¯€ë¡œ, timeout ê¸°ë°˜ ë£¨í”„
        if shell.closed:
            break
        time.sleep(0.1)

    ssh.close()


def lamp_ftp() : ## ì™„ë£Œ
    return """
    #!/bin/bash
set -e

echo "===== Rocky Linux ì„œë²„ ì´ˆê¸°í™” ìŠ¤í¬ë¦½íŠ¸ ì‹œì‘ ====="

# 0ï¸âƒ£ firewalld ë¹„í™œì„±í™”
echo "firewalld ë¹„í™œì„±í™” ì¤‘..."
systemctl stop firewalld
systemctl disable firewalld
echo "firewalld ë¹„í™œì„±í™” ì™„ë£Œ!"


# 3ï¸âƒ£ Apache ì„¤ì¹˜
echo "Apache ì„¤ì¹˜ ì¤‘..."
dnf -y install httpd
systemctl enable --now httpd
echo "Apache ì„¤ì¹˜ ë° ìë™ ì‹œì‘ ì™„ë£Œ!"

# 4ï¸âƒ£ ì‚¬ìš©ì ì›¹ ë””ë ‰í† ë¦¬ ìƒì„±
WEB_DIR="/home/heeyoung"
mkdir -p "$WEB_DIR"
chmod 755 "$WEB_DIR"
chown -R $USER:$USER "$WEB_DIR"

# SELinux í™œì„±í™” ì‹œ ì»¨í…ìŠ¤íŠ¸ ì ìš©
if sestatus | grep -q "enabled"; then
    chcon -R -t httpd_sys_content_t "$WEB_DIR"
fi
echo "ì›¹ ë””ë ‰í† ë¦¬ ìƒì„± ë° ê¶Œí•œ ì„¤ì • ì™„ë£Œ!"

# 5ï¸âƒ£ VirtualHost ì„¤ì •
VHOST_CONF="/etc/httpd/conf.d/heeyoung.conf"
cat <<EOF > "$VHOST_CONF"
<VirtualHost *:80>
    ServerName www.heeyoung.com
    DocumentRoot "$WEB_DIR"
    <Directory "$WEB_DIR">
        AllowOverride None
        Options Indexes FollowSymLinks
        Require all granted
    </Directory>
</VirtualHost>
EOF
echo "VirtualHost ì„¤ì • ì™„ë£Œ: $VHOST_CONF"

# 6ï¸âƒ£ index.html ìƒì„±
echo "qqq" > "$WEB_DIR/index.html"
echo "index.html ìƒì„± ì™„ë£Œ!"

# 7ï¸âƒ£ Apache ì¬ì‹œì‘
systemctl restart httpd
echo "Apache ì¬ì‹œì‘ ì™„ë£Œ!"

# 8ï¸âƒ£ vsftpd ì„¤ì¹˜ ë° ì‹œì‘
echo "vsftpd ì„¤ì¹˜ ì¤‘..."
dnf -y install vsftpd
systemctl enable --now vsftpd
echo "vsftpd ì„¤ì¹˜ ë° ìë™ ì‹œì‘ ì™„ë£Œ!"

echo "ğŸ‰ ëª¨ë“  ì„¤ì • ì™„ë£Œ! ì›¹ ì„œë²„ì™€ FTP ì„œë²„ê°€ ì¤€ë¹„ë˜ì—ˆìŠµë‹ˆë‹¤."

    """


bash_script = lamp_ftp()

run_remote(bash_script)
```
