```python
import paramiko
import subprocess

ip = "172.16.16.25"
user = "root"
pw = "asd123!@"
cmd = ""
partition = ""
mount_dir = ""
RAID1_MOUNT_DIR = ""

def cc(cmd):
    cli = paramiko.SSHClient()
    cli.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    cli.connect(ip, username=user, password=pw, port=22)
    stdin, stdout, stderr = cli.exec_command(cmd, get_pty=True)  # ← 여기가 핵심
    out = stdout.read().decode()
    err = stderr.read().decode()
    print(out if out else err)
    cli.close()

def pp(stdout):
    for i in stdout:
        print(i)

def install() :
    cc("dnf -y install quota")

def make_partition():
    global partition 
    partition = input("무엇을 파티션할건지 적으세요 (ex. /dev/sda): ")
    script = f'''
#!/usr/bin/expect -f
spawn fdisk {partition}
expect "Command (m for help):"
send "n\\r"
expect "Select (default p):"
send "\\r"
expect "Partition number (1-4, default 1):"
send "\\r"
expect "First sector"
send "\\r"
expect "Last sector"
send "\\r"
expect "Command (m for help):"
send "w\\r"
expect eof
'''

    # 서버에 스크립트 파일 생성 후 실행
    cc(f"echo '{script}' > /tmp/mkpart.exp")
    cc("chmod +x /tmp/mkpart.exp")
    cc("expect /tmp/mkpart.exp")
    cc("lsblk")  # 파티션 생성 확인
    ## lsblk시 sda           8:0    0  102M  0 disk로 disk 로 떠야합니다

def format_partition():
    global partition 
    partition = input("아까 생성한 파티션을 입력하세요(ex. /jupyter or /dev/sda1): ")
    cc(f'mkfs -t ext4 {partition}')
    ## lsblk시 sda           8:0    0  102M  0 part로 part 로 떠야합니다
   

## ssh는 안되는 방식 ㅠ.ㅠ 
# def permanent_mount_config():
#     mount_file = "/etc/fstab"
#     partition = input("아까 만든 새 파티션을 입력하세요 (ex. /dev/sda1): ")
#     mount_dir = input("마운트 할 경로를 입력하세요 (ex. /mnt/test): ")
#     new_line = f"{partition} {mount_dir} ext4    defaults        0 0\n"

#     with open(mount_file, "r", encoding="utf-8") as f:
#         lines = f.readlines()

#     if any(partition in line and mount_dir in line for line in lines):
#         print("이미 /etc/fstab에 존재합니다.")
#         return

#     with open(mount_file, "a", encoding="utf-8") as f:
#         f.write(new_line)
#     print("fstab에 새 항목을 추가했습니다:", new_line.strip())



def add_user() :
    # 사용자 계정 생성
    cmd = """
    # team1 계정이 존재하지 않으면 생성
    if ! id team1 &>/dev/null; then
        useradd -m -d /jupyter/team1 team1
    else
        # 이미 존재하면 홈디렉토리만 확인 또는 새로 만듦
        mkdir -p /jupyter/team1
        usermod -d /jupyter/team1 team1
    fi
    """
    cc(cmd)

# 비밀번호 설정 expect 스크립트
    script = '''
    spawn passwd team1
    expect "New password:"
    send "asd123!@\\r"
    expect "Retype new password:"
    send "asd123!@\\r"
    expect eof
    '''
    
    cc("echo '{}' > /tmp/mkuser.exp".format(script))
    cc("chmod +x /tmp/mkuser.exp")
    cc("expect /tmp/mkuser.exp")

    
def mount() :
    global mount_dir 
    mount_dir =  input("마운트 할 경로를 입력하세요 (ex. /mnt/test): ")

    cmd = f"""
    sudo mkdir -p {mount_dir}
    sudo mount -t ext4 {partition} {mount_dir}
    ls {mount_dir}
    """
    cc(cmd)

    # 명령 실행 후 출력받기
   # stdout = cc(cmd)
    # 결과를 Jupyter에서 바로 보기
 #   pp(stdout) # pp에서 에러 떠도 생기면 생긴겁니다. lost+found 터미널 들어가서서 생겻는지 직접확인하세요

def permanent_mount_config_remote():
    new_line = f"{partition} {mount_dir} ext4    defaults,usrquota        0 0\n"

    fstab_contents = cc("cat /etc/fstab")  # 문자열 반환

    cc(f"echo '{new_line}' >> /etc/fstab")
    print("fstab에 새 항목 추가함:", new_line)
    cc("systemctl daemon-reload")
    cc(f"mount -o remount {mount_dir}")
################# diskquota ###################    
def diskquota_mount() :
    # 파일 시스템 언마운트 후에 ext4 내장 쿼터 기능 활성화
    # 이부분은 루트권한이 필요합니다.. 안될시에 터미널에 직접 입력하세요.
    global mount_dir 
    mount_dir = input("diskquota를 실행할 mount 디렉토리를 입력하세요. ex. /mnt/mirror")
    cmd = f"""
    sudo bash -lc '
    umount {mount_dir}
    tune2fs -O quota {partition}
    mount -a
    quotacheck -cu {mount_dir}
    quotaon {mount_dir}
    '
    """
    cc(cmd)
def restrict_disk_and_check() :
    cc(f""" 
    sudo bash -lc '
    #edquota -u team1  <- 이건 편집하는 기능이라..
    quotaon -u {mount_dir}
    repquota {mount_dir}
    '
    """) 

################# raid1 ###################

def raid1_install():
    print(cc("dnf install -y mdadm"))
   

def raid1_mk():
    print("raid1을 생성합니다. lsblk로 남는 sata 두개가 있는지 확인하시고 이전에 파티션을 생성하지 않았는지 확인하세요.")
    yes_ = input("이전에 스냅샷을 반드시 찍어주세요. 확인 : yes , 이 문구가 다시 뜰땐 성공한것이니 no입력하세요. lsblk로 md0 생성되었는지 확인")
    if yes_ != "yes":
        print("취소되었습니다.")
        return 
    mount_disk1 = input("mount할 디스크1을 입력해주세요 ex. /dev/sda: ")
    mount_disk2 = input("mount할 디스크2를 입력해주세요 ex. /dev/sdb: ")
    cmd = f"yes | mdadm --create /dev/md0 --level=1 --raid-devices=2 {mount_disk1} {mount_disk2}"
    cc(cmd)

    print(cc("cat /proc/mdstat"))

def raid1_mk_fs_mount():
    global RAID1_MOUNT_DIR
    RAID1_MOUNT_DIR = input("mount 할 디렉토리를 입력해주세요 ex./mnt/mirror")
    cc(f"mkdir -p {RAID1_MOUNT_DIR}")
    cmd = f"""
sudo mkfs.ext4 /dev/md0
sudo mount /dev/md0 {RAID1_MOUNT_DIR}
"""
    cc(cmd)
    print("✅ RAID1 포맷 및 마운트 완료!")
    print(cc("df -h | grep md0"))

def raid1_permanent_mount_config_remote():
 
    cc(f"mkdir -p {RAID1_MOUNT_DIR}")
    new_line = f"/dev/md0    {RAID1_MOUNT_DIR}    ext4    defaults,usrquota    0    0\n"

    fstab_contents = cc("cat /etc/fstab")  # 문자열 반환

    cc(f"echo '{new_line}' >> /etc/fstab")
    print("fstab에 새 항목 추가함:", new_line)
    cc("systemctl daemon-reload")
    cc(f"mount -o remount {RAID1_MOUNT_DIR}")    
    cc("mount -a")

# 이거는 오직 /dev/sda로 sda만 파티션 하고싶을때입니다.
def only_one_partitioning():
    install()
    make_partition() # <- 이거 하고 나서 그 다음 함수 실행이 안됩니다 === 별로 끊어서 
    # 위 함수 두개 주석처리 하시고 다시 실행하세요
   #==================
    format_partition()
    add_user()
   #===================
    mount() 
    permanent_mount_config_remote()
    diskquota()
    restrict_disk_and_check()

    
# 이게 raid1 파티셔닝입니다
def raid1_partitioning():
    raid1_install()
    raid1_mk()
    raid1_mk_fs_mount()
    raid1_permanent_mount_config_remote()

def diskquota():
    diskquota_mount()
    restrict_disk_and_check()

#only_one_partitioning() <- 이거하지마세요 /dev/sda를 /dev/sda1로 하나만 파티셔닝 하는거에요.
#raid1_partitioning()
diskquota()
```
