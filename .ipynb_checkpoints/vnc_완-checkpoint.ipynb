```python
import paramiko
import time
import sys
ip = "172.16.16.122"
user = "root"
pw = "asd123!@"

def cc(cmd):
    """SSH로 명령을 실행하고 결과를 출력"""
    cli = paramiko.SSHClient()
    cli.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    cli.connect(ip, username=user, password=pw, port=22)
    stdin, stdout, stderr = cli.exec_command(cmd)
    out = stdout.read().decode()
    err = stderr.read().decode()
    cli.close()
    if err:
        print("⚠️ 오류 발생:", err)
    return out.strip()

def cc_interactive(script, timeout=30):
    cli = paramiko.SSHClient()
    cli.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    cli.connect(ip, username=user, password=pw)
    shell = cli.invoke_shell()
    shell.send(script + "\n")

    start = time.time()
    while True:
        if shell.recv_ready():
            out = shell.recv(1024).decode()
            print(out, end='')
            sys.stdout.flush()
        if time.time() - start > timeout:
            break
        time.sleep(0.1)

    shell.close()


def install_and_configure() :
    cc('dnf install tigervnc-server -y')
    cc("""
        VNC_FILE="/etc/tigervnc/vncserver.users"
        touch "$VNC_FILE"
        grep -q '^:1=root' "$VNC_FILE" || echo ':1=root' >> "$VNC_FILE"
        """)
    print("VNC 사용자 등록 완료")

    

def set_vnc_password():
    return """
    /usr/bin/expect <<'EOF'
    spawn vncpasswd
    expect "Password:"
    send "asd123!@\r"
    expect "Verify:"
    send "asd123!@\r"
    expect "Would you like to enter a view-only password (y/n)?"
    send "y\r"
    expect "Password:"
    send "asd123!@\r"
    expect "Verify:"
    send "asd123!@\r"
    expect eof
    EOF
    """
 
def install_gnome() :
    cc('dnf groupinstall -y "Server with GUI"')
    cc("systemctl set-default graphical.target")

def setup_vnc_xstartup():
    vnc_xstartup = """
    mkdir -p /root/.vnc
    cat <<'EOF' > /root/.vnc/xstartup
    #!/bin/bash
    unset SESSION_MANAGER
    unset DBUS_SESSION_BUS_ADDRESS
    exec /usr/bin/gnome-session &
    EOF
    chmod +x /root/.vnc/xstartup
    """
  
    cc(vnc_xstartup)
    # 실행 권한 부여
    cc('chmod +x /root/.vnc/xstartup')
    print("✅ VNC xstartup 설정 완료")

def vnc_service() :
    cc("systemctl daemon-reload")
    cc("systemctl enable vncserver@:1")
    cc("systemctl start vncserver@:1")
    cc("systemctl status vncserver@:1")
    
def main() :
    install_and_configure()
    cc_interactive(set_vnc_password())
    print("비번 등록완료")
    install_gnome()
    setup_vnc_xstartup()
    vnc_service()
main()

##<IP>:5901 입력하세요
```
