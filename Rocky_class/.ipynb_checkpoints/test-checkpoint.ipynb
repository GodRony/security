```python
from CC import CC
from DnsVhost import DnsVhost

cc =  CC("172.16.16.25", "root", "asd123!@")
cc.connect()
dv = DnsVhost() 


#################### HTTP , DNS ######################
#---------------------HTTP-----------------------#
## 1. http 설치, vhost 수정, 재시작 : 밑의 mod_http_R 함수 사용
def install_and_setting_http_R() :
    cc.run(dv.install_http_R())
    mod_http_and_dns_R()

## 2. vhost 수정, 재시작
def mod_http_and_dns_R():
    cc.run(dv.set_vhostconf_R())
    set_httpd_and_dns_R()
    cc.run(dv.restart_http_R())

## 3. vhost 수정, 만약 zone파일이 존재하면 vhost의 내용을 반영하고 3차도메인을 수정합니다.
def set_httpd_and_dns_R() :
    vhost_conf = dv.getter_VHOST_CONF()
    print(f"현재 당신의 vhost.conf 파일은 {vhost_conf}입니다.")
    while(True) :
        cc.run(dv.show_vhostconf_R())
        num = int(input("(1) vhost 추가하기, (0) vhost 파일 수정 종료"))
        if(num == 0) :
            print("vhost 파일 수정을 종료합니다")
            return
        cc.run(dv.mod_vhostconf_R())
        
    cc.run(dv.restart_http_R())
    cc.run(dv.set_named_conf_R())
    
    if(dv.getter_ZONE_FILE() != "") :
        cc.dv.set_zone_file_R()
        cc.set_dns_Third_domain()
#---------------------DNS-----------------------#
## 1. DNS 설치 및 ZONE 파일 수정, vhost에 따른 3차도메인 자동 업데이트
def install_and_setting_dns_R():
    cc.run(dv.install_named_R())
    cc.run(dv.set_named_conf_R())
    cc.run(dv.set_named_rfc_R())
    cc.run(dv.mk_zone_file_R())
    
    path, contents = dv.set_zone_file_R()
    cc.w_ssh(path, contents)
    path, contents = dv.set_dns_Third_domain()
    cc.a_ssh(path, contents)
    cc.run(dv.start_dns_R())
    

#install_and_setting_http_R()
install_and_setting_dns_R()
#cc.run(dv.show_vhostconf_R())
```
