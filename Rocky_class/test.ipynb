```python
from CC import CC
from DnsVhost import DnsVhost

cc =  CC("172.16.16.25", "root", "asd123!@")
cc.connect()
dv = DnsVhost() 


## http 설치, vhost 수정, 재시작 : 밑의 mod_http_R 함수 사용
def install_http_R() :
    cc.run(dv.install_http_R())
    mod_http_R()

## vhost 수정, 재시작
def mod_http_R():
    cc.run(dv.set_vhostconf_R())
    set_httpd_R()
    cc.run(dv.restart_http_R())

## vhost 수정
def set_httpd_R() :
    vhost_conf = dv.getter_VHOST_CONF()
    print(f"현재 당신의 vhost.conf 파일은 {vhost_conf}입니다.")
    while(True) :
        cc.run(dv.show_vhostconf_R())
        num = int(input("(1) vhost 추가하기, (0) vhost 파일 수정 종료"))
        if(num == 0) :
            print("vhost 파일 수정을 종료합니다")
            return
        cc.run(dv.mod_vhostconf_R())
        
    command = dv.restart_http_R()    
    cc.run(command)

install_http_R()
#cc.run(dv.show_vhostconf_R())
```
