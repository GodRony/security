```python
import paramiko
import subprocess

ip = "172.16.16.122"
user = "root"
pw = "asd123!@"
cmd = ""

def cc(cmd):
    cli = paramiko.SSHClient()
    cli.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    cli.connect(ip, username=user, password=pw, port=22)
    stdin, stdout, stderr = cli.exec_command(cmd, get_pty=True)  # ← 여기가 핵심
    out = stdout.read().decode()
    err = stderr.read().decode()
    print(out if out else err)
    cli.close()

def gnome():
    cc("sudo apt update;sudo apt install ubuntu-desktop -y")
    cc("systemctl set-default graphical.target")
def install_vnc():
    cc("apt install tigervnc-standalone-server tigervnc-xorg-extension")


def set_passwd():
    cc( """
    /usr/bin/expect <<'EOF'
    spawn vncpasswd
    expect "Password:"
    send "asd123!@\r"
    expect "Verify:"
    send "asd123!@\r"
    expect "Would you like to enter a view-only password (y/n)?"
    send "y\r"
    expect "Password:"
    send "asd123!@\r"
    expect "Verify:"
    send "asd123!@\r"
    expect eof
    EOF
    """
      )
def adduser_xrdp():
    cc("cd ~/.vnc")
    cc("adduser xrdp ssl-cert")

def setup_vnc_xstartup():
    vnc_xstartup = """
    mkdir -p /root/.vnc
    cat <<'EOF' > /root/.vnc/xstartup
    !/bin/bash
export XKL_XMODMAP_DISABLE=1
[ -x /etc/vnc/xstartup ] && exec /etc/vnc/xstartup
[ -r $HOME/.Xresources ] && xrdb $HOME/.Xresources
vncconfig -iconic &
dbus-launch --exit-with-session startplasma-x11
    EOF
    chmod +x /root/.vnc/xstartup
    """
    cc(vnc_xstartup)
    cc('chmod +x /root/.vnc/xstartup')
    print("✅ VNC xstartup 설정 완료")    

def start_vnc():
    cc("vncserver")
    
def go():
    gnome()
    install_vnc()
    set_passwd()
    adduser_xrdp()
    setup_vnc_xstartup()
    start_vnc()
    
go()
```
