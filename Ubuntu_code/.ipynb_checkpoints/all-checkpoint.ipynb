```python
#======================================================#
#                                                      #
#                  2,3ë²ˆ    D H C P                    #
#                                                      #
#======================================================#

##################### dhcp server #####################
def dhcp_install():
    """DHCP ì„œë²„ ì„¤ì¹˜"""
    cmd = """
if [ "$(id -u)" -ne 0 ]; then
    echo "ì´ ìŠ¤í¬ë¦½íŠ¸ëŠ” root ê¶Œí•œìœ¼ë¡œ ì‹¤í–‰í•´ì•¼ í•©ë‹ˆë‹¤."
    exit 1
fi
DEBIAN_FRONTEND=noninteractive apt update -y
DEBIAN_FRONTEND=noninteractive apt install isc-dhcp-server -y
"""
    cc(cmd)
    print("DHCP ì„œë²„ ì„¤ì¹˜ ì™„ë£Œ")


def set_dhcp_file():
    """DHCP ì„¤ì • íŒŒì¼ êµ¬ì„±"""
    # /etc/default/isc-dhcp-server í™•ì¸
    print(cc("cat /etc/default/isc-dhcp-server"))
    
    # ì‚¬ìš©ì ì…ë ¥
    NI = input("ìœ„ì—ì„œ í™•ì¸í•œ NIC ì´ë¦„ì„ ì…ë ¥í•´ì£¼ì„¸ìš” ex. ens33: ")
    print(cc("ip a"))
    subnet = input("ìœ„ì—ì„œ í™•ì¸í•œ subnetì„ ì…ë ¥í•˜ì„¸ìš” ex. 192.168.0.0: ")
    netmask = input("ìœ„ì—ì„œ í™•ì¸í•œ netmaskì„ ì…ë ¥í•˜ì„¸ìš” ex. 255.255.255.0: ")
    exclusion_start = input("ì œì™¸í•  ip ì‹œì‘ ì£¼ì†Œë¥¼ ì ìœ¼ì„¸ìš” ex. 192.168.0.100: ")
    exclusion_end = input("ì œì™¸í•  ip ë§ˆì§€ë§‰ ì£¼ì†Œë¥¼ ì ìœ¼ì„¸ìš” ex. 192.168.0.200: ")
    gateway = input("ê²Œì´íŠ¸ì›¨ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš” ex. 192.168.0.1: ")
    domain = input("ë„ë©”ì¸ ì„œë²„ë¥¼ ì…ë ¥í•˜ì„¸ìš” ex. 8.8.8.8: ")
    pool = input("poolì˜ ì´ë¦„ì„ ì§€ì •í•˜ì„¸ìš” ex. 'example.local': ")

    # NIC ì—…ë°ì´íŠ¸
    ch_NI_cmd = f"""
sed -i 's|^INTERFACESv4=.*$|INTERFACESv4="{NI}"|' /etc/default/isc-dhcp-server
"""
    cc(ch_NI_cmd)
    print("/etc/default/isc-dhcp-server íŒŒì¼ì„ í•´ë‹¹ NIë¡œ ë³€ê²½í–ˆìŠµë‹ˆë‹¤.")

    # dhcpd.conf ì—…ë°ì´íŠ¸
    add_dhcp_file_cmd = f"""
echo '
subnet {subnet} netmask {netmask} {{
    range {exclusion_start} {exclusion_end};
    option routers {gateway};
    option subnet-mask {netmask};
    option domain-name-servers {domain}, 8.8.8.8;
    option domain-name "{pool}";
}}
' >> /etc/dhcp/dhcpd.conf
"""
    cc(add_dhcp_file_cmd)
    print("/etc/dhcp/dhcpd.conf íŒŒì¼ ì—…ë°ì´íŠ¸ ì™„ë£Œ")

def dhcp_restart():
    cmd = """
systemctl restart isc-dhcp-server
systemctl enable isc-dhcp-server
systemctl status isc-dhcp-server
    """
    cc(cmd)
    
##################### dhcp client #####################

def dhcp_request():
    """DHCP ìš”ì²­ ë° ì¸í„°í˜ì´ìŠ¤ í™•ì¸"""
    cmd = """
apt install isc-dhcp-client -y
dhclient -v
ip a
systemctl restart isc-dhcp-server
systemctl enable isc-dhcp-server
systemctl status isc-dhcp-server
"""
    cc(cmd)
    print("DHCP ìš”ì²­ ì™„ë£Œ ë° ì„œë²„ ì‹¤í–‰ í™•ì¸")



###################### ì‹¤í–‰ ########################
def dhcp_server_in():
    dhcp_install()
    set_dhcp_file()
    dhcp_request()

def dhcp_client_in():
    dhcp_request()

```

```python
#======================================================#
#                                                      #
#                  4ë²ˆ    HTTP , FTP                   #
#                                                      #
#======================================================#

def set_ufw():
    """ë°©í™”ë²½ ìƒíƒœ í™•ì¸ ë° ì¤‘ì§€"""
    print("ğŸ”¥ ë°©í™”ë²½ ì„¤ì • ì¤‘...")
    cmd = """
systemctl stop ufw
systemctl disable ufw
echo 'UFW ë°©í™”ë²½ ì¤‘ì§€ ì™„ë£Œ'
"""
    cc(cmd)


def install_package():
    """Apache ì„¤ì¹˜"""
    print("ğŸ“¦ Apache ì„¤ì¹˜ ì¤‘...")
    cmd = """
apt update -y
apt install -y apache2
"""
    cc(cmd)

##################### HTTP #####################
def start_apache():
    """Apache ì‹¤í–‰"""
    print("ğŸš€ Apache ì‹¤í–‰ ì¤‘...")
    cmd = """
systemctl enable apache2
systemctl restart apache2
systemctl status apache2 --no-pager
"""
    cc(cmd)


def set_user_dir_and_vhost_conf():
    """ì‚¬ìš©í•  ì›¹ ë””ë ‰í† ë¦¬ ì…ë ¥"""
    global WEB_DOCUMENT_ROOT
    WEB_DOCUMENT_ROOT = input("ì‚¬ìš©í•  ì›¹ ë””ë ‰í† ë¦¬ë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: /home/team1 ë˜ëŠ” /var/www/young): ").strip()
    cc(f"mkdir -p {WEB_DOCUMENT_ROOT} && chmod 755 {WEB_DOCUMENT_ROOT}")
    print(f"âœ… ì›¹ ë¬¸ì„œ ë£¨íŠ¸: {WEB_DOCUMENT_ROOT}")

def set_apache2_conf() :
    cmd = f"""
    sed -i 's|<Directory /var/www/>|<Directory {WEB_DOCUMENT_ROOT}>|' {APACHE2_CONF}
    """
    cc(cmd)

def set_vhostconf():
    """ê°€ìƒ í˜¸ìŠ¤íŠ¸ ì„¤ì • íŒŒì¼ ìƒì„±"""
    global VHOST_CONF, SERVER_NAME, ERROR_LOG, CUSTOM_LOG

    vhost_name = input("VHost ì„¤ì • íŒŒì¼ì„ ë§Œë“¤ê² ìŠµë‹ˆë‹¤ ì›í•˜ëŠ” ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: team1.conf): ")
    VHOST_CONF = f"{VHOST_DIR}/{vhost_name}"

    SERVER_NAME = input("ServerNameì„ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: test.local): ")
    ERROR_LOG = input("ErrorLog ê²½ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: /var/log/apache2/team1_error.log): ")
    CUSTOM_LOG = input("CustomLog ê²½ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: /var/log/apache2/team1_access.log): ")

    print(f" {VHOST_CONF} ìƒì„± ì¤‘...")

    cmd = f"""
cat <<EOF > {VHOST_CONF}
<VirtualHost *:80>
    ServerAdmin webmaster@localhost
    ServerName {SERVER_NAME}
    DocumentRoot {WEB_DOCUMENT_ROOT}

    <Directory {WEB_DOCUMENT_ROOT}>
        AllowOverride None
        Options Indexes FollowSymLinks
        Require all granted
    </Directory>

    ErrorLog {ERROR_LOG}
    CustomLog {CUSTOM_LOG} combined
</VirtualHost>
EOF

a2ensite {vhost_name}
a2dissite 000-default.conf
systemctl reload apache2
mkdir -p {ERROR_LOG}; mkdir -p {CUSTOM_LOG};
echo ' VirtualHost ì„¤ì • ì™„ë£Œ: {VHOST_CONF}'
"""
    cc(cmd)


def mk_index():
    """ê¸°ë³¸ index.html ìƒì„±"""
    print("ğŸ“„ index.html ìƒì„± ì¤‘...")
    cmd = f"echo '<h1>This is index.html for Test by mk_index func</h1>' > {WEB_DOCUMENT_ROOT}/index.html"
    print(f"{WEB_DOCUMENT_ROOT} ê²½ë¡œì— íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤")
    cc(cmd)


def restart_apa():
    """Apache ì¬ì‹œì‘"""
    print("ğŸ” Apache ì¬ì‹œì‘ ì¤‘...")
    cc("systemctl restart apache2 && echo 'Apache ì¬ì‹œì‘ ì™„ë£Œ'")

##################### FTP #####################

def install_ftp():
    cc("apt install -y vsftpd")

def start_ftp():
    cc("systemctl enable vsftpd; systemctl start vsftpd")

###################### ì‹¤í–‰ ########################
def http_in():
    """ì „ì²´ ìë™í™” ì‹¤í–‰"""
    set_ufw()
    install_package()
    start_apache()
    set_user_dir_and_vhost_conf()
    set_apache2_conf()
    set_vhostconf()
    mk_index()
    restart_apa()
    install_ftp()
    start_ftp()
```

```python
#==========================================================================#
#                                                                          #
#  12 , 13, 14ë²ˆ Make RAID1, Make Partition and Foramt & Mount , diskquota #
#                                                                          #
#==========================================================================#

MK_PARTITION = ""
PARTITION = ""
MOUNT_DIR = ""
USER_NAME = ""
RAID1_MOUNT_DIR = "" 

def install() :
    cc("apt -y install quota")
    cc("apt install -y expect")

############### Make Partition and Foramt & Mount #############

def make_partition():
    print("ê¸°ë³¸ì ì¸ íŒŒí‹°ì…˜ í•˜ë‚˜ë§Œ ìƒì„±í•©ë‹ˆë‹¤.")
    global MK_PARTITION
    MK_PARTITION = input("ë¬´ì—‡ì„ íŒŒí‹°ì…˜í• ê±´ì§€ ì ìœ¼ì„¸ìš” (ex. /dev/sda): ")
    script = f'''
#!/usr/bin/expect -f
spawn fdisk {MK_PARTITION}
expect "Command (m for help):"
send "n\\r"
expect "Select (default p):"
send "\\r"
expect "Partition number (1-4, default 1):"
send "\\r"
expect "First sector"
send "\\r"
expect "Last sector"
send "\\r"
expect "Command (m for help):"
send "w\\r"
expect eof
'''

    # ì„œë²„ì— ìŠ¤í¬ë¦½íŠ¸ íŒŒì¼ ìƒì„± í›„ ì‹¤í–‰
    cc(f"echo '{script}' > /tmp/mkpart.exp")
    cc("chmod +x /tmp/mkpart.exp")
    cc("expect /tmp/mkpart.exp")
    cc("lsblk")  # íŒŒí‹°ì…˜ ìƒì„± í™•ì¸
    ## lsblkì‹œ sda           8:0    0  102M  0 diskë¡œ disk ë¡œ ë– ì•¼í•©ë‹ˆë‹¤


def format_partition():
    global PARTITION
    PARTITION = input("ì•„ê¹Œ ë§Œë“  ìƒˆ íŒŒí‹°ì…˜ì„ ì…ë ¥í•˜ì„¸ìš” (ex. /dev/sda1): ")
    cc(f'mkfs -t ext4 {PARTITION}')
    ## lsblkì‹œ sda           8:0    0  102M  0 partë¡œ part ë¡œ ë– ì•¼í•©ë‹ˆë‹¤

def add_user() :
    # ì‚¬ìš©ì ê³„ì • ìƒì„±
    global USER_NAME
    USER_NAME = input("ìƒì„±í•  ì‚¬ìš©ì ê³„ì •ëª…ì„ ì…ë ¥í•˜ì„¸ìš” ")
    
    cmd = f"""
    # team1 ê³„ì •ì´ ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ ìƒì„±
    if ! id team1 &>/dev/null; then
        useradd -m -d /jupyter/{USER_NAME} {USER_NAME}
    else
        # ì´ë¯¸ ì¡´ì¬í•˜ë©´ í™ˆë””ë ‰í† ë¦¬ë§Œ í™•ì¸ ë˜ëŠ” ìƒˆë¡œ ë§Œë“¦
        mkdir -p /jupyter/{USER_NAME}
        usermod -d /jupyter/{USER_NAME} {USER_NAME}
    fi
    """
    cc(cmd)

# ë¹„ë°€ë²ˆí˜¸ ì„¤ì • expect ìŠ¤í¬ë¦½íŠ¸
    script = f'''
    spawn passwd {USER_NAME}
    expect "New password:"
    send "asd123!@\\r"
    expect "Retype new password:"
    send "asd123!@\\r"
    expect eof
    '''
    
    cc("echo '{}' > /tmp/mkuser.exp".format(script))
    cc("chmod +x /tmp/mkuser.exp")
    cc("expect /tmp/mkuser.exp")

    
def mount() :
    global MOUNT_DIR
    MOUNT_DIR = input("ë§ˆìš´íŠ¸ í•  ê²½ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš” (ex. /mnt/test): ")

    cmd = f"""
    sudo mkdir -p {MOUNT_DIR}
    sudo mount -t ext4 {PARTITION} {MOUNT_DIR}
    ls {MOUNT_DIR}
    """
    cc(cmd)

    # ëª…ë ¹ ì‹¤í–‰ í›„ ì¶œë ¥ë°›ê¸°
   # stdout = cc(cmd)
    # ê²°ê³¼ë¥¼ Jupyterì—ì„œ ë°”ë¡œ ë³´ê¸°
 #   pp(stdout) # ppì—ì„œ ì—ëŸ¬ ë– ë„ ìƒê¸°ë©´ ìƒê¸´ê²ë‹ˆë‹¤. lost+found í„°ë¯¸ë„ ë“¤ì–´ê°€ì„œì„œ ìƒê²»ëŠ”ì§€ ì§ì ‘í™•ì¸í•˜ì„¸ìš”

def permanent_mount_config_remote():
    new_line = f"{PARTITION} {MOUNT_DIR} ext4    defaults,usrquota        0 0\n"

    fstab_contents = cc("cat /etc/fstab")  # ë¬¸ìì—´ ë°˜í™˜

    cc(f"echo '{new_line}' >> /etc/fstab")
    print("fstabì— ìƒˆ í•­ëª© ì¶”ê°€í•¨:", new_line)
    cc("systemctl daemon-reload")
    cc(f"mount -o remount {MOUNT_DIR}")

################# diskquota ###################
def diskquota_mount() :
    global mount_dir 
    cc("dnf -y install quota")
    mount_dir = input("diskquotaë¥¼ ì‹¤í–‰í•  mount ë””ë ‰í† ë¦¬ë¥¼ ì…ë ¥í•˜ì„¸ìš”. ex. /mnt/mirror")
    
    cmd = f"""
    sudo bash -lc '
    umount {mount_dir}
    tune2fs -O quota {mount_dir}
    mount -a
    quotacheck -cu {mount_dir}
    quotaon {mount_dir}
    '
    """
    cc(cmd)
def restrict_disk_and_check() :
    cc(f""" 
    sudo bash -lc '
    #edquota -u team1  <- ì´ê±´ í¸ì§‘í•˜ëŠ” ê¸°ëŠ¥ì´ë¼..
    quotaon -u {mount_dir}
    repquota {mount_dir}
    '
    """) 

################# raid1 ###################

def raid1_install():
    cc("apt install -y mdadm")
   

def raid1_mk():
    print("raid1ì„ ìƒì„±í•©ë‹ˆë‹¤. lsblkë¡œ ë‚¨ëŠ” sata ë‘ê°œê°€ ìˆëŠ”ì§€ í™•ì¸í•˜ì‹œê³  ì´ì „ì— íŒŒí‹°ì…˜ì„ ìƒì„±í•˜ì§€ ì•Šì•˜ëŠ”ì§€ í™•ì¸í•˜ì„¸ìš”.")
    yes_ = input("ì´ì „ì— ìŠ¤ëƒ…ìƒ·ì„ ë°˜ë“œì‹œ ì°ì–´ì£¼ì„¸ìš”. í™•ì¸ : yes , ì´ ë¬¸êµ¬ê°€ ë‹¤ì‹œ ëœ°ë• ì„±ê³µí•œê²ƒì´ë‹ˆ noì…ë ¥í•˜ì„¸ìš”. lsblkë¡œ md0 ìƒì„±ë˜ì—ˆëŠ”ì§€ í™•ì¸")
    if yes_ != "yes":
        print("ì·¨ì†Œë˜ì—ˆìŠµë‹ˆë‹¤.")
        return 
    mount_disk1 = input("mountí•  ë””ìŠ¤í¬1ì„ ì…ë ¥í•´ì£¼ì„¸ìš” ex. /dev/sda: ")
    mount_disk2 = input("mountí•  ë””ìŠ¤í¬2ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš” ex. /dev/sdb: ")
    cmd = f"yes | mdadm --create /dev/md0 --level=1 --raid-devices=2 {mount_disk1} {mount_disk2}"
    cc(cmd)

    print(cc("cat /proc/mdstat"))

def raid1_mk_fs_mount():
    global RAID1_MOUNT_DIR
    RAID1_MOUNT_DIR = input("mount í•  ë””ë ‰í† ë¦¬ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš” ex./mnt/mirror")
    cc(f"mkdir -p {RAID1_MOUNT_DIR}")

    make_partition()
    cc(f"sudo mkfs.ext4 /dev/md0p1")
    cc(f"sudo mount /dev/md0 {RAID1_MOUNT_DIR}")
    print("âœ… RAID1 í¬ë§· ë° ë§ˆìš´íŠ¸ ì™„ë£Œ!")
    print(cc("df -h | grep md0"))



def raid1_permanent_mount_config_remote():
    
    cc(f"mkdir -p {RAID1_MOUNT_DIR}")
    new_line = f"/dev/md0p1    {RAID1_MOUNT_DIR}    ext4    defaults,usrquota    0    0\n"

    fstab_contents = cc("cat /etc/fstab")  # ë¬¸ìì—´ ë°˜í™˜
    
    cc(f"echo '{new_line}' >> /etc/fstab")
    print("fstabì— ìƒˆ í•­ëª© ì¶”ê°€í•¨:", new_line)
    cc("systemctl daemon-reload")
    cc(f"mount -o remount {RAID1_MOUNT_DIR}")    
    cc("mount -a")
    
# ì´ê±°ëŠ” ì˜¤ì§ /dev/sdaë¡œ sdaë§Œ íŒŒí‹°ì…˜ í•˜ê³ ì‹¶ì„ë•Œì…ë‹ˆë‹¤.
def mk_partition_in():
    install()
   #==================
    make_partition()
    format_partition()
    add_user()
   #===================
    mount() 
    permanent_mount_config_remote()
    diskquota()
    restrict_disk_and_check()
    
def mk_raid1_in():
    install()
    raid1_install()
    raid1_mk()
    raid1_mk_fs_mount()
    raid1_permanent_mount_config_remote()

def diskquota():
    install()
    diskquota_mount()
    restrict_disk_and_check()
```

```python
#======================================================#
#                                                      #
#        15 , 16ë²ˆ DNS ì„¤ì¹˜ ë° ì‹¤í–‰, DNS DATA ê´€ë¦¬        #
#                                                      #
#======================================================#

ZONE_FILE_NAME = ""
ZONE_NAME = ""

########### DNS ì„¤ì¹˜ ë° ì‹¤í–‰ ##############

def install():
    print("=== BIND9 ì„¤ì¹˜ ===")
    pp(cc("apt-get update -y"))
    pp(cc("apt-get install -y bind9 bind9-utils bind9-dnsutils"))
    pp(cc("systemctl status bind9"))

def mk_zone_file():
    global ZONE_FILE_NAME
    ZONE_FILE_NAME = input("ìƒì„±í•  zone íŒŒì¼ì„ ì…ë ¥í•´ì£¼ì„¸ìš” (ex. test.com.zone): ")
    print(f"db.local(example) íŒŒì¼ì„ ë³µì‚¬í•´ {ZONE_FILE_NAME} zone íŒŒì¼ì„ ìƒì„±í•©ë‹ˆë‹¤.")
    cc(f"cp -ap /etc/bind/db.local /etc/bind/{ZONE_FILE_NAME}")
    print(f"âœ… zone íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤: /etc/bind/{ZONE_FILE_NAME}")

def set_zone_file():
    global ZONE_FILE_NAME
    print("zone fileì˜ ë‚´ìš©ì„ ìˆ˜ì •í•˜ê² ìŠµë‹ˆë‹¤.")
    domain = input("ì‚¬ìš©í•  ë„ë©”ì¸ ì£¼ì†Œ (ex. ns.test.com): ")
    ipaddr = input("í˜„ì¬ ì‚¬ìš©í•˜ê³  ìˆëŠ” DNSì˜ ip (ip a í™•ì¸) (ex. 172.16.16.100): ")
    contents = f"""$TTL    604800
@       IN      SOA     {domain}. root.adminemail.com. (
                              2         ; Serial
                         604800         ; Refresh
                          86400         ; Retry
                        2419200         ; Expire
                         604800 )       ; Negative Cache TTL
;
@       IN      NS      {domain}.
ns       IN      A       {ipaddr}
@       IN      AAAA    ::1
"""
    w_ssh(f"/etc/bind/{ZONE_FILE_NAME}", contents)
    print("âœ… zone file ë‚´ìš© ìˆ˜ì • ì™„ë£Œ")

def set_named_conf_options():
    print("bind9 ì„¤ì • íŒŒì¼ì„ ìˆ˜ì •í•˜ê² ìŠµë‹ˆë‹¤. (ê²½ë¡œ : /etc/bind/named.conf.options)")
    contents = """options {
        directory "/var/cache/bind";
        dnssec-validation auto;
        recursion yes;
        allow-query { any; };
        listen-on { any; };
        listen-on-v6 { any; };
};
"""
    conf_path = "/etc/bind/named.conf.options"
    cc(f"cp -ap {conf_path} {conf_path}.bak")
    w_ssh(conf_path, contents)
    print("âœ… named.conf.options ìˆ˜ì • ì™„ë£Œ")

def set_named_conf_local():
    global ZONE_NAME, ZONE_FILE_NAME
    print("bind9 ì„¤ì • íŒŒì¼ì„ ìˆ˜ì •í•˜ê² ìŠµë‹ˆë‹¤. (ê²½ë¡œ : /etc/bind/named.conf.local)")
    ZONE_NAME = input("zoneì˜ ì´ë¦„ì„ ì…ë ¥í•´ì£¼ì„¸ìš” (ex. test.com): ")
    contents = f"""
zone "{ZONE_NAME}" {{
    type master;
    file "/etc/bind/{ZONE_FILE_NAME}";
}};
"""
    a_ssh("/etc/bind/named.conf.local", contents)
    print(f"âœ… zone '{ZONE_NAME}' ì¶”ê°€ ì™„ë£Œ (/etc/bind/named.conf.local)")

def start_dns():
    pp(cc("systemctl restart bind9"))
    pp(cc("systemctl status bind9"))

########### dns_dataë¥¼ í™œìš©í•´ dnsê´€ë¦¬ ##############

def create_dns_file():
    file_name = "dns_data.txt"
    target_path = "/root/"
    cc(f"touch {target_path}{file_name}")
    print(f"âœ… {target_path}{file_name} ìƒì„± ì™„ë£Œ")

def info():
    name = input("ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš” (ë°˜ì˜ë˜ì§€ ì•ŠëŠ” ì •ë³´ì…ë‹ˆë‹¤. í™•ì¸ìš©) : ")
    domain_id = input("ì‚¬ìš©í•  3ì°¨ ë„ë©”ì¸ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš” (ex. ww1, ww2): ")
    ip = input("IPë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")
    data = f"{name} {domain_id} {ip}\n"
    return data

def dns_mod():
    file_path = "/root/dns_data.txt"
    while True:
        lines = r_ssh(file_path)
        lines_list = lines.splitlines()

        print("\n==== í˜„ì¬ íŒŒì¼ ë‚´ìš© ====")
        for i, line in enumerate(lines_list, 1):
            print(f"{i}: {line}")
        print("======================\n")

        input_num = input("ìˆ˜ì •(1), ì‚­ì œ(2), ì¶”ê°€(3), ì¢…ë£Œ(0): ")

        if input_num == "0":
            print("í”„ë¡œê·¸ë¨ ì¢…ë£Œ")
            break
        elif input_num == "1":
            line_num = int(input("ìˆ˜ì •í•  ì¤„ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
            if 1 <= line_num <= len(lines_list):
                new_text = info()
                lines_list[line_num - 1] = new_text
                w_ssh(file_path, "\n".join(lines_list))
        elif input_num == "2":
            line_num = int(input("ì‚­ì œí•  ì¤„ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
            if 1 <= line_num <= len(lines_list):
                lines_list.pop(line_num - 1)
                w_ssh(file_path, "\n".join(lines_list))
        elif input_num == "3":
            new_text = info()
            a_ssh(file_path, new_text + "\n")
        else:
            print("ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤.")


def overwrite_dns():
 #   zone_file = f"/etc/bind/{ZONE_FILE_NAME}"
    zone_file = "/etc/bind/test.com.zone"
    data_file = "/root/dns_data.txt"

    data_lines = r_ssh(data_file).splitlines()
    zone_lines = r_ssh(zone_file).splitlines()

    header = zone_lines[:12]
    new_records = []
    
    for line in data_lines:
        parts = line.strip().split()
        if len(parts) >= 3:
            name, domain, ip = parts[:3]
            new_records.append(f"{domain}\tIN\tA\t{ip}")

    new_zone_content = header + new_records
    new_zone_content_str = "\n".join(new_zone_content) + "\n"

    print(new_zone_content_str)
    a_ssh(zone_file, new_zone_content_str)
    print("Zone íŒŒì¼ì´ ì„±ê³µì ìœ¼ë¡œ ê°±ì‹ ë˜ì—ˆìŠµë‹ˆë‹¤.")


################# ì‹¤í–‰ë¶€ ##################
def dns_in,():
    install()
    mk_zone_file()
    ##############
    set_zone_file()
    set_named_conf_options()
    set_named_conf_local()
    start_dns()

def dns_data_txt_in():
    create_dns_file()
    dns_mod()
    overwrite_dns()

```
