```python
import paramiko

# ì„œë²„ ì ‘ì† ì •ë³´
ip = "172.16.16.24"
user = "root"
pw = "asd123!@"

# ê³ ì •ëœ ê²½ë¡œ
HTTPD_CONF = "/etc/apache2/sites-available/000-default.conf"
VHOST_DIR = "/etc/apache2/sites-available"
APACHE2_CONF = "/etc/apache2/apache2.conf"

# ë™ì ìœ¼ë¡œ ë°›ì„ ê°’
WEB_DOCUMENT_ROOT = ""
VHOST_CONF = ""
SERVER_NAME = ""
ERROR_LOG = ""
CUSTOM_LOG = ""


def cc(cmd):
    """SSHë¡œ ëª…ë ¹ì–´ë¥¼ ì‹¤í–‰í•˜ê³  ê²°ê³¼ ì¶œë ¥"""
    cli = paramiko.SSHClient()
    cli.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    cli.connect(ip, username=user, password=pw, port=22)
    stdin, stdout, stderr = cli.exec_command(cmd, get_pty=True)
    out = stdout.read().decode()
    err = stderr.read().decode()
    print(out if out else err)
    cli.close()


def set_ufw():
    """ë°©í™”ë²½ ìƒíƒœ í™•ì¸ ë° ì¤‘ì§€"""
    print("ğŸ”¥ ë°©í™”ë²½ ì„¤ì • ì¤‘...")
    cmd = """
systemctl stop ufw
systemctl disable ufw
echo 'UFW ë°©í™”ë²½ ì¤‘ì§€ ì™„ë£Œ'
"""
    cc(cmd)


def install_package():
    """Apache ì„¤ì¹˜"""
    print("ğŸ“¦ Apache ì„¤ì¹˜ ì¤‘...")
    cmd = """
apt update -y
apt install -y apache2
"""
    cc(cmd)


def start_apache():
    """Apache ì‹¤í–‰"""
    print("ğŸš€ Apache ì‹¤í–‰ ì¤‘...")
    cmd = """
systemctl enable apache2
systemctl restart apache2
systemctl status apache2 --no-pager
"""
    cc(cmd)


def set_user_dir_and_vhost_conf():
    """ì‚¬ìš©í•  ì›¹ ë””ë ‰í† ë¦¬ ì…ë ¥"""
    global WEB_DOCUMENT_ROOT
    WEB_DOCUMENT_ROOT = input("ì‚¬ìš©í•  ì›¹ ë””ë ‰í† ë¦¬ë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: /home/team1 ë˜ëŠ” /var/www/young): ").strip()
    cc(f"mkdir -p {WEB_DOCUMENT_ROOT} && chmod 755 {WEB_DOCUMENT_ROOT}")
    print(f"âœ… ì›¹ ë¬¸ì„œ ë£¨íŠ¸: {WEB_DOCUMENT_ROOT}")

def set_apache2_conf() :
    cmd = f"""
    sed -i 's|<Directory /var/www/>|<Directory {WEB_DOCUMENT_ROOT}>|' {APACHE2_CONF}
    """
    cc(cmd)

def set_vhostconf():
    """ê°€ìƒ í˜¸ìŠ¤íŠ¸ ì„¤ì • íŒŒì¼ ìƒì„±"""
    global VHOST_CONF, SERVER_NAME, ERROR_LOG, CUSTOM_LOG

    vhost_name = input("VHost ì„¤ì • íŒŒì¼ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: team1.conf): ")
    VHOST_CONF = f"{VHOST_DIR}/{vhost_name}"

    SERVER_NAME = input("ServerNameì„ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: test.local): ")
    ERROR_LOG = input("ErrorLog ê²½ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: /var/log/apache2/team1_error.log): ")
    CUSTOM_LOG = input("CustomLog ê²½ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: /var/log/apache2/team1_access.log): ")

    print(f"ğŸ› ï¸ {VHOST_CONF} ìƒì„± ì¤‘...")

    cmd = f"""
cat <<EOF > {VHOST_CONF}
<VirtualHost *:80>
    ServerAdmin webmaster@localhost
    ServerName {SERVER_NAME}
    DocumentRoot {WEB_DOCUMENT_ROOT}

    <Directory {WEB_DOCUMENT_ROOT}>
        AllowOverride None
        Options Indexes FollowSymLinks
        Require all granted
    </Directory>

    ErrorLog {ERROR_LOG}
    CustomLog {CUSTOM_LOG} combined
</VirtualHost>
EOF

a2ensite {vhost_name}
a2dissite 000-default.conf
systemctl reload apache2
mkdir -p {ERROR_LOG}; mkdir -p {CUSTOM_LOG};
echo 'âœ… VirtualHost ì„¤ì • ì™„ë£Œ: {VHOST_CONF}'
"""
    cc(cmd)


def mk_index():
    """ê¸°ë³¸ index.html ìƒì„±"""
    print("ğŸ“„ index.html ìƒì„± ì¤‘...")
    cmd = f"echo '<h1>This is index.html for Test by mk_index func</h1>' > {WEB_DOCUMENT_ROOT}/index.html"
    print(f"{WEB_DOCUMENT_ROOT} ê²½ë¡œì— íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤")
    cc(cmd)


def restart_apa():
    """Apache ì¬ì‹œì‘"""
    print("ğŸ” Apache ì¬ì‹œì‘ ì¤‘...")
    cc("systemctl restart apache2 && echo 'Apache ì¬ì‹œì‘ ì™„ë£Œ'")

def install_ftp():
    cc("apt install -y vsftpd")

def start_ftp():
    cc("systemctl enable vsftpd; systemctl start vsftpd")

def go():
    """ì „ì²´ ìë™í™” ì‹¤í–‰"""
#    set_ufw()
#    install_package()
#    start_apache()
    set_user_dir_and_vhost_conf()
    set_apache2_conf()
#    set_vhostconf()
#    mk_index()
#    restart_apa()
    print("ğŸ‰ ëª¨ë“  ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!")

    install_ftp()
    start_ftp()
if __name__ == "__main__":
    go()
```
