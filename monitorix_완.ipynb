```python
import paramiko

# SSH 명령 실행 함수
def ssh(ip, user, pw, cmd):
    cli = paramiko.SSHClient()
    cli.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    cli.connect(ip, username=user, password=pw, port=22)

    stdin, stdout, stderr = cli.exec_command(cmd)
    print(stdout.read().decode('utf-8', errors='ignore'))
    print(stderr.read().decode('utf-8', errors='ignore'))
    
    cli.close()

# Rocky 서버 접속 정보
ip = "172.16.16.122"
user = "root"
pw = "asd123!@"

# Monitorix 설치
def download():
    cmds = [
        "dnf install -y monitorix perl perl-FindBin",
        "dnf install -y perl-CGI perl-HTTP-Server-Simple perl-rrdtool perl-Config-General perl-LWP-Protocol-https perl-LWP-Protocol-http10 perl-libwww-perl",
        "dnf install -y perl-Config-General perl-IO-Socket-SSL perl-DBI perl-XML-Simple perl-CGI perl-libwww-perl perl-HTTP-Server-Simple",
        "dnf install -y perl perl-FindBin"
    ]
    for c in cmds:
        print(f"=== 실행 중: {c} ===")
        ssh(ip, user, pw, c)

# Monitorix 재시작
def restart():
    ssh(ip, user, pw, "systemctl restart monitorix")

# Monitorix 상태 확인
def status():
    ssh(ip, user, pw, "systemctl status monitorix")

# 전체 실행
def main():
    print("=== Monitorix 설치 시작 ===")
    download()
    print("=== 서비스 재시작 ===")
    restart()
    print("=== 서비스 상태 확인 ===")
    status()

main()
```
