```python
import pymysql

ip = "172.16.16.100"
port = "3306"
id_ = "root"
pw = "asd123!@"

db_name = "test_db3"
user_name = "test_user3"

db = pymysql.connect(host=ip, user=id_,passwd=pw,charset='utf8') # 커넥터 생성

cur = db.cursor() # cursor : 접속한 상태를 가리킴.

cmd = "show databases" # DBMS에는 ; 반드시 붙여야하지만 자동으로 여기선 붙여줌

cur.execute(cmd) # 배치 방식 

cur.execute(f"CREATE DATABASE {db_name}")
cur.execute(f"CREATE USER '{user_name}'@'{ip}' IDENTIFIED BY '{pw}'")
cur.execute(f"GRANT ALL PRIVILEGES ON {db_name}.* TO '{user_name}'@'{ip}'")
cur.execute("FLUSH PRIVILEGES")

rows = cur.fetchall() # DMS에게 commit 전송하고 그 후에 반환되는 값을 rows에 초기화한다.

for i in rows:
    print(i)

cur.close()
```
