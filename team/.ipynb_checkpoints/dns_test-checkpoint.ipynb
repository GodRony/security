```python
import paramiko
import time
import sys
import os

ip = "172.16.16.100"
user = "root"
pw = "asd123!@"

def cc(cmd):
    """SSHë¡œ ëª…ë ¹ì„ ì‹¤í–‰í•˜ê³  ê²°ê³¼ë¥¼ ì¶œë ¥"""
    cli = paramiko.SSHClient()
    cli.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    cli.connect(ip, username=user, password=pw, port=22)
    stdin, stdout, stderr = cli.exec_command(cmd)
    out = stdout.read().decode()
    err = stderr.read().decode()
    cli.close()
    if err:
        print("âš ï¸ ì˜¤ë¥˜ ë°œìƒ:", err)
    return out.strip()

def install() :
    cc( 'echo "=== install ===" ')
    cc("dnf -y install bind bind-utils")
    cc("systemctl status named")

def modify_named_conf_file() :
    #!/bin/bash
    cmd = r"""
    NAMED_CONF="/etc/named.conf"
    sed -i 's/listen-on[[:space:]]*port 53[[:space:]]*{[[:space:]]*127\.0\.0\.1;[[:space:]]*};/listen-on port 53 { any; };/' "$NAMED_CONF"
    sed -i 's/allow-query[[:space:]]*{[[:space:]]*localhost;[[:space:]]*};/allow-query { any; };/' "$NAMED_CONF"
    """
    print(cc(cmd))
    print("âœ… named.conf ìˆ˜ì • ì™„ë£Œ")

def modify_named_rfc_file() :

    cmd = r"""
    NAMED_CONF="/etc/named.rfc1912.zones"

    cat >> "$NAMED_CONF" <<'EOF'

zone "team1.com" IN {
        type master;
        file "team1.com";
        allow-update { none; };
};
EOF
    """
    print(cc(cmd))
    print("âœ… zone 'team1.com' ì„¤ì •ì´ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.")

def add_namezone() :
    cc("cp -ap /var/named/named.localhost /var/named/team1.com.zone")

## ì—¬ê¸° ë¶€ë¶„ë¶€í„°ëŠ” ì›ê²©ìœ¼ë¡œëŠ” ëª»í•¨.. ##

def edit_default_zone():
    zone_file = "/var/named/team1.com.zone"  # ì‹¤ì œ ê²½ë¡œë¡œ ë³€ê²½í•˜ì„¸ìš”
    
    # ë°±ì—…
    with open(zone_file, "r") as f:
        lines = f.readlines()
    
    with open(zone_file + ".bak", "w") as f:
        f.writelines(lines)
    print("ğŸ“¦ ë°±ì—… ìƒì„± ì™„ë£Œ:", zone_file + ".bak")
    
    # ìˆ˜ì •
    new_lines = []
    for line in lines:
        if "@ rname.invalid." in line:
            line = "@" + "\tIN SOA  ns.team1.com. ns.team1.com. (\n"
        elif "NS" in line and "@" in line:
            line = "        NS      ns.team1.com.\n"
        elif "A" in line and "127.0.0.1" in line:
            line = "        A       172.16.16.100\n"
        new_lines.append(line)
    
    # ë®ì–´ì“°ê¸°
    with open(zone_file, "w") as f:
        f.writelines(new_lines)
    
    print("âœ… ê¸°ë³¸ ì¡´ íŒŒì¼ ìˆ˜ì • ì™„ë£Œ:", zone_file)

#####################################################
##             dns_data.txt ìƒì„±ì½”ë“œ                ##
#####################################################


## 1. dns ìƒì„± íŒŒì¼ 
def create_dns_file() :
    file_name = "dns_data.txt"
    target_path = "../output/"
    command1 = "touch %s%s"%(target_path, file_name)
    a = os.popen(command1)
    data = a.read()

# ================================  #

## 2. dns íŒŒì¼ ì¶”ê°€ ìƒì„± ì½”ë“œ
def info():
    name = input("ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
    domain_id = input("ì‚¬ìš©í•  ë„ë©”ì¸ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
    ip = input("IPë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")

    # ì›í•˜ëŠ” í¬ë§·ìœ¼ë¡œ ë¬¸ìì—´ ìƒì„±
    data = f"{name} {domain_id} {ip}"
    return data

def dns_mod() :
    file_path = "../output/dns_data.txt"
    while True:
    # íŒŒì¼ ì½ê¸°
        with open(file_path, "r") as f:
            content = f.read()
    
        lines = content.splitlines()
        print("\n==== í˜„ì¬ íŒŒì¼ ë‚´ìš© ====")
        for i, line in enumerate(lines, 1):
            print(f"{i}: {line}")
        print("======================\n")
    
        # ì‚¬ìš©ì ì…ë ¥
        input_num = input("ìˆ˜ì •(1), ì‚­ì œ(2), ì¶”ê°€(3), ì¢…ë£Œ(0) ì›í•˜ì‹œëŠ” ë™ì‘ì˜ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")
    
        if input_num == "0":
            print("í”„ë¡œê·¸ë¨ ì¢…ë£Œ")
            break
    
        elif input_num == "1":  # ìˆ˜ì •
            line_num = int(input("ìˆ˜ì •í•  ì¤„ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
            if 1 <= line_num <= len(lines):
                # info() í•¨ìˆ˜ í˜¸ì¶œí•´ì„œ ìƒˆë¡œìš´ ë°ì´í„° ë°›ê¸°
                new_text = info()
                lines[line_num - 1] = new_text
                with open(file_path, "w") as f:
                    f.write("\n".join(lines))
            else:
                print("ì˜ëª»ëœ ì¤„ ë²ˆí˜¸ì…ë‹ˆë‹¤.")
        
        elif input_num == "2":  # ì‚­ì œ
            line_num = int(input("ì‚­ì œí•  ì¤„ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
            if 1 <= line_num <= len(lines):
                lines.pop(line_num - 1)
                # ë¹ˆ ì¤„ ì œê±°
                lines = [line for line in lines if line.strip() != ""]
                with open(file_path, "w") as f:
                    f.write("\n".join(lines))
            else:
                print("ì˜ëª»ëœ ì¤„ ë²ˆí˜¸ì…ë‹ˆë‹¤.")
    
        elif input_num == "3":  # ì¶”ê°€
            new_text = info()
            with open(file_path, "a") as f:
                f.write("\n" + new_text)
    
        else:
            print("ì˜ëª»ëœ ì„ íƒì…ë‹ˆë‹¤. 0~3 ì‚¬ì´ì˜ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”.")

## dns íŒŒì¼ì„ ë®ì–´ì“°ê¸°
def overwrite_dns() :
    zone_file = "/var/named/team1.com.zone"
    data_file = "../output/dns_data.txt"
    
    data_lines = []
    zone_lines = []
    
    
    # ì·¨í•©ëœ íŒŒì¼ ì½ê¸°
    with open(data_file, "r", encoding="utf-8") as d:
        data_lines = d.readlines()
    
    # Zone íŒŒì¼ ì½ê¸°
    with open(zone_file, "r", encoding="utf-8") as db:
        zone_lines = db.readlines()
    
    # Zone íŒŒì¼ì˜ 12ë²ˆì§¸ ì¤„ê¹Œì§€ ìœ ì§€í•˜ê³  ì´í›„ëŠ” ì‚­ì œ
    header = zone_lines[:16]
    
    # ì·¨í•©ëœ ë°ì´í„°ë¥¼ ê·œì¹™ì— ë§ê²Œ ë³€í™˜
    new_records = []
    for line in data_lines:
        parts = line.strip().split()
        if len(parts) >= 3:
            name, c3, ip = parts[:3]
            new_records.append(f"{c3}\tIN\tA\t{ip}\n\n")
    
    # í•©ì¹˜ê¸°
    new_zone_content = header + new_records
    
    # Zone íŒŒì¼ ë®ì–´ì“°ê¸°
    with open(zone_file, "w", encoding="utf-8") as f:
        f.writelines(new_zone_content)
    
    print("Zone íŒŒì¼ì´ ì„±ê³µì ìœ¼ë¡œ ê°±ì‹ ë˜ì—ˆìŠµë‹ˆë‹¤.")

    

def init() :
#    install()
#    modify_named_conf_file()
#    modify_named_rfc_file()
    add_namezone()
    create_dns_file()
    edit_default_zone()

def mod_zone() :
    dns_mod()
    overwrite_dns()

init()
mod_zone()
```
