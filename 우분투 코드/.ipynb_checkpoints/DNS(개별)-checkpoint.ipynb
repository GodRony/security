```python
import paramiko
import os

ip = "172.16.27.175"
user = "root"
pw = "asd123!@"
cmd = ""

def cc(cmd):
    cli = paramiko.SSHClient()
    cli.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    cli.connect(ip, username=user, password=pw, port=22)
    (stdin, stdout, stderr) = cli.exec_command(cmd)
    out = stdout.read().decode()
    err = stderr.read().decode()
    cli.close()
    return out + err

def pp(stdout):
    for i in stdout.splitlines():
        print(i.strip())

###################################
#            기능 함수 영역
###################################

def install():
    print(pp(cc('echo "=== BIND9 설치 ==="')))
    pp(cc("apt-get update -y"))
    pp(cc("apt-get install -y bind9 bind9-utils bind9-dnsutils"))
    pp(cc("systemctl status bind9"))

def modify_named_conf_file():
    print("🔧 named.conf.options 수정 중...")
    cmd = r"""
    CONF_FILE="/etc/bind/named.conf.options"
    sed -i 's/listen-on-v6 { any; };/\/\/listen-on-v6 { any; };/g' "$CONF_FILE"
    sed -i '/options {/a \    listen-on port 53 { any; };' "$CONF_FILE"
    sed -i '/allow-query {/c\    allow-query { any; };' "$CONF_FILE"
    """
    print(pp(cc(cmd)))
    print("✅ named.conf.options 수정 완료")

def modify_named_rfc_file():
    print("🔧 named.conf.local 수정 중...")
    cmd = r"""
    CONF_FILE="/etc/bind/named.conf.local"
    cat >> "$CONF_FILE" <<'EOF'

zone "team1.com" {
        type master;
        file "/var/lib/bind/team1.com.zone";
};
EOF
    """
    print(pp(cc(cmd)))
    print("✅ zone 'team1.com' 설정이 추가되었습니다.")

def add_namezone():
    print("📄 team1.com.zone 파일 생성 중...")
    cmd = """
    cp /etc/bind/db.local /var/lib/bind/team1.com.zone
    chown bind:bind /var/lib/bind/team1.com.zone
    """
    print(pp(cc(cmd)))
    print("✅ 기본 존 파일 생성 완료")

# 이하 로컬에서만 동작 ###################################여기 실행이 안돼 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
def edit_default_zone():
    zone_file = "/var/lib/bind/team1.com.zone"
    with open(zone_file, "r") as f:
        lines = f.readlines()
    with open(zone_file + ".bak", "w") as f:
        f.writelines(lines)
    print("📦 백업 생성 완료:", zone_file + ".bak")

    new_lines = []
    for line in lines:
        if "@ IN SOA" in line:
            line = "@   IN SOA  ns.team1.com. admin.team1.com. (\n"
        elif "NS" in line and "localhost." in line:
            line = "        NS      ns.team1.com.\n"
        elif "A" in line and "127.0.0.1" in line:
            line = "ns      IN      A       172.16.27.175\n"
        new_lines.append(line)
    with open(zone_file, "w") as f:
        f.writelines(new_lines)
    print("✅ 기본 존 파일 수정 완료:", zone_file)
###################################################################################################
def create_dns_file():
    file_name = "dns_data.txt"
    target_path = "../output/"
    os.makedirs(target_path, exist_ok=True)
    os.system(f"touch {target_path}{file_name}")
    print(f"✅ {target_path}{file_name} 생성 완료")

def info():
    name = input("이름을 입력하세요: ")
    domain_id = input("사용할 도메인 이름을 입력하세요: ")
    ip = input("IP를 입력하세요: ")
    data = f"{name} {domain_id} {ip}"
    return data

def dns_mod():
    file_path = "../output/dns_data.txt"
    while True:
        with open(file_path, "r") as f:
            content = f.read()
        lines = content.splitlines()
        print("\n==== 현재 파일 내용 ====")
        for i, line in enumerate(lines, 1):
            print(f"{i}: {line}")
        print("======================\n")
        input_num = input("수정(1), 삭제(2), 추가(3), 종료(0): ")

        if input_num == "0":
            print("프로그램 종료")
            break
        elif input_num == "1":
            line_num = int(input("수정할 줄 번호를 입력하세요: "))
            if 1 <= line_num <= len(lines):
                new_text = info()
                lines[line_num - 1] = new_text
                with open(file_path, "w") as f:
                    f.write("\n".join(lines))
        elif input_num == "2":
            line_num = int(input("삭제할 줄 번호를 입력하세요: "))
            if 1 <= line_num <= len(lines):
                lines.pop(line_num - 1)
                with open(file_path, "w") as f:
                    f.write("\n".join(lines))
        elif input_num == "3":
            new_text = info()
            with open(file_path, "a") as f:
                f.write("\n" + new_text)
        else:
            print("잘못된 입력입니다.")

def overwrite_dns():
    zone_file = "/var/lib/bind/team1.com.zone"
    data_file = "../output/dns_data.txt"

    with open(data_file, "r", encoding="utf-8") as d:
        data_lines = d.readlines()
    with open(zone_file, "r", encoding="utf-8") as db:
        zone_lines = db.readlines()

    header = zone_lines[:16]
    new_records = []
    for line in data_lines:
        parts = line.strip().split()
        if len(parts) >= 3:
            name, domain, ip = parts[:3]
            new_records.append(f"{domain}\tIN\tA\t{ip}\n\n")
    new_zone_content = header + new_records
    with open(zone_file, "w", encoding="utf-8") as f:
        f.writelines(new_zone_content)
    print("✅ Zone 파일이 성공적으로 갱신되었습니다.")

###################################
# 초기 실행 함수
###################################

def init():
    install()
    modify_named_conf_file()
    modify_named_rfc_file()
    add_namezone()
    create_dns_file()
    edit_default_zone()

def mod_zone():
    dns_mod()
    overwrite_dns()

init()
mod_zone()
```
