```python
import paramiko
import os

ip = "192.168.0.17"
user = "root"
pw = "asd123!@"
cmd = ""
ZONE_FILE_NAME = ""
ZONE_NAME = ""

def cc(cmd):
    cli = paramiko.SSHClient()
    cli.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    cli.connect(ip, username=user, password=pw, port=22)
    (stdin, stdout, stderr) = cli.exec_command(cmd)
    out = stdout.read().decode()
    err = stderr.read().decode()
    cli.close()
    return out + err

def setup_samba():
    print("smb를 설치합니다")
    cc("apt update && apt install -y samba smbclient cifs-utils")

    # 2. 설치 확인
    out, _ = run("dpkg -l | grep samba")
    for line in out.splitlines():
        if "samba" in line:
            print("삼바 서버 설치 완료")
        if "smbclient" in line:
            print("삼바 클라이언트 설치 완료")

    # 3. SMB 계정 생성 및 비밀번호 설정
    run("useradd -s /usr/sbin/nologin smb")
    PASSWORD = "asd123!@"
    expect_cmd = f'''
expect -c "
set timeout 10
spawn smbpasswd -a smb
expect \\"New SMB password:\\"
send \\"{PASSWORD}\\r\\"
expect \\"Retype new SMB password:\\"
send \\"{PASSWORD}\\r\\"
expect eof
"
'''
    run(expect_cmd)

    # 4. Samba 계정 확인
    out, _ = run("pdbedit --list")
    print(out)

    # 5. 공유 디렉터리 생성
    run("mkdir -p /share/smb")
    out, _ = run("ls /share")
    print(out)

    # 6. smb.conf 설정
    out, _ = run("cat /etc/samba/smb.conf")
    if "[share]" not in out:
        run("cd /etc/samba && cp -ap smb.conf smb.conf.bak")
        run("""echo "[share]
comment = Kong's SMB
hosts allow = 172.16.0.0/16
path = /share/smb
read only = No
valid users = smb
write list = smb
" >> /etc/samba/smb.conf""")
    else:
        print("이미 설정이 있어서 설정파일 수정 안했어요!!!")

    # 7. 권한 설정
    run("chmod 777 /share && chown smb:smb /share/smb")

    # 8. 서비스 시작 및 상태 확인
    out, _ = run("systemctl restart smbd && systemctl status smbd --no-pager")
    if "active (running)" in out:
        print("smbd 실행 성공")

    out, _ = run("systemctl restart nmbd && systemctl status nmbd --no-pager")
    if "active (running)" in out:
        print("nmbd 실행 성공")

    cli.close()
    print("Samba 설치 및 설정 완료!")
```
